## 1. Objet et méthode de la veille

Bien que le sujet du projet nous oriente explicitement vers l'API de traduction de
Google, nous avons choisi de ne pas nous limiter à ce seul service et de mener une
comparaison sérieuse avec ses principales alternatives.

Cette démarche poursuit trois objectifs :

- **Valider** que Google est effectivement adapté à notre besoin au regard d'autres options du marché.
- **Identifier des solutions de secours** en cas de difficulté technique ou administrative.
- **Justifier nos choix** à la soutenance.

Notre méthode s'est organisée en quatre étapes : formulation du besoin technique,
définition des critères de comparaison, recensement des services candidats, et étude
approfondie de chacun à partir de sa documentation officielle et de comparatifs tiers
publiés au premier semestre 2026.

## 2. Expression du besoin technique

Le besoin fonctionnel principal est de traduire un texte fourni par un utilisateur,
depuis une langue source vers une langue cible, et de renvoyer la traduction sous
une forme exploitable par notre code Python.

Le besoin secondaire est la détection automatique de la langue source.

**Contraintes techniques retenues :**

- Application codée en Python : bibliothèque cliente officielle souhaitable.
- Volumes faibles : quelques milliers à dizaines de milliers de caractères au total.
- Pas de budget : le service doit fonctionner dans son palier gratuit.
- Latence raisonnable pour ne pas dégrader l'expérience utilisateur.
- Couverture linguistique : langues européennes principales + chinois, arabe, russe, japonais.

## 3. Critères de comparaison retenus

À partir de ce besoin, nous avons défini sept critères que nous appliquons
uniformément à chaque service candidat :

1. **Couverture linguistique** — nombre et nature des langues prises en charge.
2. **Qualité de la traduction** — évaluée à partir de retours et comparatifs publics.
3. **Palier gratuit** — volume mensuel utilisable sans payer.
4. **Friction d'inscription** — notamment la nécessité d'une carte bancaire.
5. **Documentation et SDK Python** — qualité de la doc, existence d'un client officiel.
6. **Stabilité du service** — historique, disponibilité, pérennité.
7. **Détection automatique de langue** — disponible nativement ou à compléter par une autre solution.

## 4. Service 1 — Google Cloud Translation

Service de traduction commercialisé par Google dans son écosystème Cloud.
Mentionné explicitement par le sujet du projet.

| Critère | Évaluation |
| --- | --- |
| Couverture | 130+ langues *(source : BuildMVPFast)* |
| Qualité | Bonne sur toutes les paires, référence pour les langues à faibles ressources |
| Palier gratuit | 500 000 caractères / mois *(source : Costgoat)* |
| CB requise | Oui *(source : GitHub)* |
| SDK Python | Oui — `google-cloud-translate` officiel |
| Stabilité | Excellente, +15 ans en production |
| Détection de langue | Oui, endpoint dédié |

**Exemple d'appel en Python** (nécessite un compte Google Cloud configuré) :

```python
# Google Cloud Translation — example call
# Requires: pip install google-cloud-translate
# Requires: GOOGLE_APPLICATION_CREDENTIALS env var pointing to a service account JSON

from google.cloud import translate_v2 as translate

# The client picks up the credentials path from the environment automatically
client = translate.Client()

result = client.translate(
    "Bonjour le monde",
    source_language="fr",
    target_language="en",
)

print(result["translatedText"])
# Expected output: "Hello world"

# Note: this cell is shown for documentation purposes.
# It cannot be executed without a configured Google Cloud account.
```

## 5. Service 2 — DeepL API

Éditeur allemand spécialisé dans la traduction automatique, réputé pour la qualité
de ses traductions sur les langues européennes.

| Critère | Évaluation |
| --- | --- |
| Couverture | ~33 langues *(source : Langbly)* |
| Qualité | Excellente sur langues européennes majeures |
| Palier gratuit | 500 000 caractères / mois sur le tier API Free *(source : NeuraPlusAI)* |
| CB requise | Non pour le tier API Free *(source : Bluente)* |
| SDK Python | Oui — `deepl` officiel |
| Stabilité | Très bonne, entreprise établie depuis 2017 |
| Détection de langue | Oui, explicite ou implicite |

**Exemple d'appel en Python** (nécessite une clé DeepL gratuite) :

```python
# DeepL API — example call
# Requires: pip install deepl

import deepl

# Sign up at https://www.deepl.com/pro-api to get a free key (no credit card)
DEEPL_AUTH_KEY = "your_deepl_auth_key_here"

translator = deepl.Translator(DEEPL_AUTH_KEY)

result = translator.translate_text(
    "Bonjour le monde",
    source_lang="FR",
    target_lang="EN-US",
)

print(result.text)
# Expected output: "Hello world"

# Note: this cell is shown for documentation purposes.
# It cannot be executed without a valid DeepL API key.
```

## 6. Service 3 — Microsoft Azure Translator

Service de traduction de Microsoft dans son écosystème Azure AI Services.

| Critère | Évaluation |
| --- | --- |
| Couverture | 100+ langues *(source : Bluente)* |
| Qualité | Honnête, légèrement en retrait de Google et DeepL |
| Palier gratuit | 2 000 000 caractères / mois (tier F0) *(source : Point of AI)* |
| CB requise | Oui (compte Azure) |
| SDK Python | Oui — `azure-ai-translation-text` |
| Stabilité | Excellente, soutenue par Microsoft |
| Détection de langue | Oui, endpoint dédié |

**Exemple d'appel en Python** via l'API REST (nécessite un compte Azure) :

```python
# Microsoft Azure Translator — example call via REST API
# Requires: pip install requests

import requests

AZURE_KEY = "your_azure_subscription_key_here"
AZURE_REGION = "westeurope"  # adjust to your resource region
ENDPOINT = "https://api.cognitive.microsofttranslator.com/translate"

headers = {
    "Ocp-Apim-Subscription-Key": AZURE_KEY,
    "Ocp-Apim-Subscription-Region": AZURE_REGION,
    "Content-Type": "application/json",
}
params = {
    "api-version": "3.0",
    "from": "fr",
    "to": "en",
}
body = [{"text": "Bonjour le monde"}]

response = requests.post(ENDPOINT, params=params, headers=headers, json=body)
data = response.json()
print(data[0]["translations"][0]["text"])
# Expected output: "Hello world"

# Note: this cell is shown for documentation purposes.
# It cannot be executed without a configured Azure account.
```

## 7. Service 4 — MyMemory Translation API

Service de traduction basé sur une mémoire de traduction collaborative, combinée à
des modèles de traduction automatique en complément. **Service retenu pour notre
projet** (voir section 10).

| Critère | Évaluation |
| --- | --- |
| Couverture | 100+ langues *(source : Tossom)* |
| Qualité | Variable selon la disponibilité de traductions humaines dans la base |
| Palier gratuit | 5 000 car./jour anonyme, 50 000 mots/jour avec email *(source : Langbly)* |
| CB requise | Non — aucune inscription requise |
| SDK Python | Non officiel — usage direct de `requests` |
| Stabilité | Correcte, moins éprouvée que les majors |
| Détection de langue | Via `langpair=autodetect|target` (extraction depuis la réponse) |

**Exemple d'appel en Python** — cette cellule est **directement exécutable**,
aucune clé n'est requise :

```python
# MyMemory Translation API — example call
# Requires: pip install requests
# No API key needed for the anonymous tier (5,000 chars/day per IP)

import requests

MYMEMORY_URL = "https://api.mymemory.translated.net/get"

params = {
    "q": "Bonjour le monde",
    "langpair": "fr|en",
    # "de": "your_email@example.com"  # Optional: lifts quota to ~50,000 words/day
}

response = requests.get(MYMEMORY_URL, params=params, timeout=10)
data = response.json()

print(data["responseData"]["translatedText"])
# Expected output: "Hello world"

# Bonus: the match score (0 to 1) indicates translation confidence
print(f"Match score: {data['responseData']['match']}")
```

## 8. Service 5 — LibreTranslate

Initiative open source qui propose une API de traduction auto-hébergeable, basée
sur la bibliothèque Argos Translate.

| Critère | Évaluation |
| --- | --- |
| Couverture | ~30 langues *(source : Coda One)* |
| Qualité | La plus faible des options comparées *(source : Awesome F/OSS)* |
| Palier gratuit | Illimité en auto-hébergement ; instance publique principale désormais payante |
| CB requise | Non en auto-hébergement |
| SDK Python | Oui — `libretranslate` |
| Stabilité | Projet actif, instances publiques parfois indisponibles |
| Détection de langue | Oui, endpoint `/detect` dédié |

**Note importante :** depuis 2023, l'instance officielle `libretranslate.com` requiert
une clé d'API payante. L'usage gratuit nécessite soit un mirror communautaire (dont la
disponibilité est variable), soit l'auto-hébergement via Docker.

**Exemple d'appel en Python** :

```python
# LibreTranslate — example call
# Requires: pip install requests
# For free usage, either self-host (docker run libretranslate/libretranslate)
# or use a community mirror. The official libretranslate.com now requires
# a paid API key.

import requests

LIBRETRANSLATE_URL = "http://localhost:5000/translate"  # self-hosted default
# Or use a community mirror: "https://translate.terraprint.co/translate"
# Or the paid official endpoint with a key

payload = {
    "q": "Bonjour le monde",
    "source": "fr",
    "target": "en",
    "format": "text",
    # "api_key": "your_libretranslate_api_key_here"  # if using a paid endpoint
}

response = requests.post(LIBRETRANSLATE_URL, json=payload, timeout=10)
data = response.json()
print(data["translatedText"])
# Expected output: "Hello world"

# Note: this cell requires either a local LibreTranslate instance running
# on port 5000, or an accessible community mirror.
```

## 10. Choix retenu et justification

À l'issue de cette comparaison, notre choix initial s'est porté sur **Google Cloud
Translation**, en raison de sa mention explicite dans le sujet du projet, de sa
couverture linguistique étendue, et de la robustesse de son infrastructure.

En cours de projet, des contraintes administratives sur la création du compte Google
Cloud (validation bancaire bloquante au-delà de quarante-huit heures) nous ont conduits
à activer notre solution de repli. C'est ici que le travail de veille préalable s'est
révélé décisif : nous avions identifié et évalué plusieurs alternatives accessibles
sans inscription contraignante, ce qui nous a permis de basculer sans repartir de zéro.

Le service finalement retenu est **MyMemory Translation API**, pour les raisons suivantes :

1. **Immédiatement utilisable** sans inscription ni carte bancaire — résout complètement
   notre contrainte d'accès.
2. **Couverture de plus de 100 langues**, suffisante pour notre interface.
3. **API REST simple** consommée via la bibliothèque standard `requests`.
4. **Palier gratuit confortable** : 5 000 caractères/jour anonyme, extensible à
   ~50 000 mots/jour avec un email facultatif.

Nous avons écarté les autres alternatives pour les raisons suivantes :

- **Microsoft Azure Translator** offrait le palier gratuit le plus généreux, mais
  présentait la même contrainte de validation bancaire que Google.
- **DeepL** a une couverture linguistique trop restreinte (33 langues).
- **LibreTranslate** présente une qualité de traduction insuffisante au regard
  de nos exigences, et son instance publique principale est devenue payante.

Cette expérience illustre l'utilité concrète d'une veille technique menée avec rigueur
en amont du développement : disposer d'un panel évalué de solutions alternatives nous
a permis de réagir en quelques heures à un blocage qui aurait pu coûter plusieurs jours.
